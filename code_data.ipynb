```python
#2020년도 건축년도별 주택 데이터를 통해서 2011~2020까지의 데이터를 채우기

house_age = pd.read_csv('./2020/35_2020년_건축년도별주택.txt',sep='^',header=None)
house_age.sort_values([1,2],inplace=True)

house_age.set_index(1,inplace=True)
house_age.drop([0],axis=1,inplace=True)
house_age
```

```python
group = house_age.groupby(2)
group_frames = [group_data for _,group_data in group]
merge_df = pd.concat(group_frames,axis=1)
merge_df.columns = [i for i in range(merge_df.shape[1])]

for column in merge_df.columns:
    if column%2 == 0:
        f_index = merge_df[column].first_valid_index()
        f_value = str(merge_df.loc[f_index,column])
    elif column%2 == 1:
        merge_df = merge_df.rename(columns={merge_df.columns[column]: f_value})
    
for column in merge_df.columns:
    if type(column) == int:
        merge_df.drop(column,axis=1,inplace=True)

merge_df
```

```python
for index,row in merge_df.iterrows():
    if row['ho_yr_999'] != 0.0:
        row.fillna(0,inplace=True)
    else:
        row['ho_yr_999'] = 1

merge_df
```

```python
#각 집계구 데이터를 분기별로 4개로 복사해서 각각 1,2,3,4 (겨울, 봄, 여름, 가을)
#각 계절에 어느 달을 넣을 것 인지는 화재발생데이터에서도 적용되는 겨울 12,1,2월 봄 3,4,5월 여름 6,7,8월 가을 9,10,11월로 입력

for i in range(10):
    df1 = merged_list[i].copy()
    df2 = merged_list[i].copy()
    df3 = merged_list[i].copy()
    df4 = merged_list[i].copy()
    df1['season'] = 1 #겨울
    df2['season'] = 2 #봄
    df3['season'] = 3 #여름
    df4['season'] = 4 #가을
    merged_list[i] = pd.concat([df1,df2,df3,df4],axis = 0)

merged_list[0]
```

```python
#m_list는 각 연도별 집계구 데이터에 날씨 데이터를 합친 것을 저장하는 리스트, 0~9 인덱스가 2011~2020년으로 배정되어 있음
#알고리즘은 먼저 날씨데이터에서 각 계절별 월 정보만 뽑아냄 -- 3개의 월별 날씨 정보를 하나의 행으로 합치기(c1을 먼저 만들어 놓고 나머지는 만들면서 c1에 합치는 방법 사용) -- 날씨 데이터 처리된 열에 계절 데이터 추가,필요없는 열 제거 -- 기존 집계구 데이터와 도시이름,연도,계절을 맞추어서 병합

m_list = []
for j in range(10):
    s1 = sum[((sum['month'] >= 1) & (sum['month'] < 3)) | (sum['month'] == 12)] #겨울
    #월별 날씨 한 행으로 합치기
    c1 = pd.DataFrame(pd.concat([s1.iloc[i].add_suffix(i) for i in range(0,3)],axis = 0)).T
    for k in range(1,140):
        c2 = pd.DataFrame(pd.concat([s1.iloc[i].add_suffix(i%3) for i in range(3*k,3*k+3)],axis = 0)).T
        c3 = pd.concat([c1,c2],axis=0)
        c1 = c3
    #계절데이터 추가
    c1['season'] = 1
    #필요없는 열 제거
    c1.drop(c1.columns[[10,11,20,21]],axis=1,inplace=True)
    #기존 집계구 데이터와 병합
    m1 = pd.merge(merged_list[j][merged_list[j]['season'] == 1],c1,left_on = ['si_name','year','season'],right_on = ['city0','year0','season'],how = 'left')
        
    s2 = sum[(sum['month'] >= 3) & (sum['month'] < 6)] #봄
    c1 = pd.DataFrame(pd.concat([s2.iloc[i].add_suffix(i) for i in range(0,3)],axis = 0)).T
    for k in range(1,140):
        c2 = pd.DataFrame(pd.concat([s2.iloc[i].add_suffix(i%3) for i in range(3*k,3*k+3)],axis = 0)).T
        c3 = pd.concat([c1,c2],axis=0)
        c1 = c3
    c1['season'] = 2
    c1.drop(c1.columns[[10,11,20,21]],axis=1,inplace=True)
    m2 = pd.merge(merged_list[j][merged_list[j]['season'] == 2],c1,left_on = ['si_name','year','season'],right_on = ['city0','year0','season'],how = 'left')
        
    s3 = sum[(sum['month'] >= 6) & (sum['month'] < 9)] #
    c1 = pd.DataFrame(pd.concat([s3.iloc[i].add_suffix(i) for i in range(0,3)],axis = 0)).T
    for k in range(1,140):
        c2 = pd.DataFrame(pd.concat([s3.iloc[i].add_suffix(i%3) for i in range(3*k,3*k+3)],axis = 0)).T
        c3 = pd.concat([c1,c2],axis=0)
        c1 = c3
    c1['season'] = 3
    c1.drop(c1.columns[[10,11,20,21]],axis=1,inplace=True)
    m3 = pd.merge(merged_list[j][merged_list[j]['season'] == 3],c1,left_on = ['si_name','year','season'],right_on = ['city0','year0','season'],how = 'left')
        
    s4 = sum[(sum['month'] >= 9) & (sum['month'] < 12)] #가을
    c1 = pd.DataFrame(pd.concat([s4.iloc[i].add_suffix(i) for i in range(0,3)],axis = 0)).T
    for k in range(1,140):
        c2 = pd.DataFrame(pd.concat([s4.iloc[i].add_suffix(i%3) for i in range(3*k,3*k+3)],axis = 0)).T
        c3 = pd.concat([c1,c2],axis=0)
        c1 = c3
    c1['season'] = 4
    c1.drop(c1.columns[[10,11,20,21]],axis=1,inplace=True)
    m4 = pd.merge(merged_list[j][merged_list[j]['season'] == 4],c1,left_on = ['si_name','year','season'],right_on = ['city0','year0','season'],how = 'left')

    #계절별로 처리된 데이터를 다시 하나로 합쳐서 연도별 집계구 데이터로 만들기
    m_list.append(pd.concat([m1,m2,m3,m4],axis = 0))

```
